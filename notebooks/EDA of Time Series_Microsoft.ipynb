```python
import pandas_datareader as pdr
import pandas as pd
from datetime import datetime
import yfinance as yf
```

```python
df = yf.download('MSFT')
df
```

```python
df.plot()
```

```python
df['High'].plot(figsize=(12, 4))
```

```python
# get the maximum price of the share from 2017 to 2022

df.loc['2017-01-01':'2022-12-31'].max()
```

```python
df.loc['2017-01-01':'2022-12-31'].max().plot()
```

```python
# get the data of the highest price of the stock

max_price = df['Close'].max()
highest_price_data = df[df['Close'] == max_price]
```

```python
max_price
```

```python
highest_price_data
```

```python
min_price = df['Close'].min()
lowest_price_data = df[df['Close'] == min_price]
min_price
```

```python
lowest_price_data
```

```python
# rule = 'A' means year end frequency(means the end of the year data you are basically going to start)

df.resample(rule='A')
```

```python
df.resample(rule='A').min()
```

```python
df.resample(rule='A').max()
```

```python
df.resample(rule='A').max()['Open'].plot()
```

```python
# quaterly start frequency

df.resample(rule='QS').max()
```

```python
# business end frequency

df.resample(rule='BA').max()
```

```python
df['Open'].resample(rule='A').mean().plot(kind='bar')
```

```python
# getting monthlt mean

df['Open'].resample(rule='M').mean()
```

```python
df['High'].rolling(10).mean()
```
