## [Simple Moving Average](https://www.investopedia.com/terms/s/sma.asp)

```python
df['Open'].plot(figsize=(12, 4))
```

```python
df['Open:10 days rolling'] = df['Open'].rolling(window=10, min_periods=1).mean()
```

```python
df[['Open', 'Open:10 days rolling']].plot(figsize=(12,6))
```

```python
df[['Open', 'Open:10 days rolling']].plot(xlim=['2020-01-01', '2021-01-01'], figsize=(12,6))
```

```python
df['Open:30 days rolling'] = df['Open'].rolling(window=30, min_periods=1).mean()
df['Open:50 days rolling'] = df['Open'].rolling(window=50, min_periods=1).mean()
```

```python
df[['Open', 'Open:10 days rolling', 'Open:30 days rolling', 'Open:50 days rolling']].plot(xlim=['2020-01-01', '2021-01-01'], figsize=(12,6))
```

* But there is a disadvantage in Simple Moving Average:
    - we give similar importance to all the data but time series data give more weight to the Recent Data.

## [Cummulative Moving Average](https://www.automateexcel.com/stats/cumulative-rolling-average/)

```python
# Using Expanding Function

df['Open'].expanding().mean().plot(figsize=(10,5))
```

## [Exponential Moving Average](https://www.investopedia.com/terms/e/ema.asp)

```python
# EMA = ((close value - previous EMA) * multiplier) + previous EMA, also multiplier = (2/(rolling window size + 1))

# smoothing factor = 0.1
df['EMA_0.1'] = df['Open'].ewm(alpha=0.1, adjust=False).mean()
```

```python
df[['Open', 'EMA_0.1']].plot(xlim=['2020-01-01', '2021-01-01'], figsize=(12,6))
```

* Without any lag we can able to see the entire Exponential Moving Average

* Advantage - removing lags and we are also trying to find out the exponential moving average which is pretty much better than                 the Simple Moving Average. 

```python
# changing smoothing factor = 0.3

df['EMA_0.3'] = df['Open'].ewm(alpha=0.3, adjust=False).mean()
```

```python
df[['Open', 'EMA_0.1', 'EMA_0.3']].plot(xlim=['2020-01-01', '2021-01-01'], figsize=(12,6))
```

```python
df['EMA_5days'] = df['Open'].ewm(span=5).mean()
```

```python
df[['Open', 'EMA_0.1', 'EMA_5days']].plot(figsize=(12,6))
```
